```python
# I am assuming for now that all of these imports are necissary
# (will figure each out later)

#imports
import numpy as np
import matplotlib.pyplot as plt
import astropy
from astropy.io import ascii
from StringIO import StringIO
from matplotlib.colors import LogNorm
import pylab
from matplotlib.ticker import NullFormatter
import numpy.ma as ma
np.set_printoptions(threshold='nan')
import pickle
from collections import Counter  

# Import the python3 print function
from __future__ import print_function
```

```python
# I don't know what this does
# I think it reads the catalog or something
WDS_Master2 =astropy.io.ascii.read('WDS_CSV_cat.txt', delimiter =',', guess=False, Reader =ascii.NoHeader, fill_values=[('.', '-999'), ('', '-999')])
```

```python
# I think this is converting the read catalog into named variables for readability
Coords = WDS_Master2['col1']
Discoverer_and_number = WDS_Master2['col2']
Components = WDS_Master2['col3']
Date_first = WDS_Master2['col4']
Date_last = WDS_Master2['col5']
N_obs = WDS_Master2['col6']
Pos_angle_first = WDS_Master2['col7']
Pos_angle_last = WDS_Master2['col8']
Sep_first = WDS_Master2['col9']
Sep_last = WDS_Master2['col10']
Pri_mag = WDS_Master2['col11']
Sec_mag = WDS_Master2['col12']
Spectral_type = WDS_Master2['col13']
Pri_RA_prop_motion = WDS_Master2['col14']
Pri_DEC_prop_motion = WDS_Master2['col15']
Sec_RA_prop_motion = WDS_Master2['col16']
Sec_DEC_prop_motion = WDS_Master2['col17']
D_num = WDS_Master2['col18']
Notes = WDS_Master2['col19']
Arcsec_coords = WDS_Master2['col20']
```

```python
# I think this is taking out the part that KAPOW can see 

#Create Subsets of the Data
Pri_mag_lim = Pri_mag[np.argwhere((Pri_mag>0.0) & (Pri_mag <20.0))]
Sep_last_lim = Sep_last[np.argwhere((Sep_last>0.0) & (Sep_last <20.0))]
```

```python
# I think this is taking out the part that is interesting for KAPOW? 

# Derive secondary data parameters

# Fill in None values with numbers 
Pri_mag_g = ma.filled(Pri_mag,[-999])
Sec_mag_g = ma.filled(Sec_mag,[-999])

# Convert to array 
Pri_mag_gg = np.asarray(Pri_mag_g, dtype = 'float_' )
Sec_mag_gg = np.asarray(Sec_mag_g, dtype = 'float_' )

# This is the important part 
# Interesting separation and magnitude range
Delta_mag = np.subtract(Pri_mag_gg, Sec_mag_gg)
Delta_mag_lim = Delta_mag[np.argwhere((Delta_mag<0.0) & (Delta_mag>-20.0) & (Sep_last>0.0) & (Sep_last <20.0))]
Sep_last_lim2 = Sep_last[np.argwhere((Delta_mag<0.0) & (Delta_mag>-20.0) & (Sep_last>0.0) & (Sep_last <20.0))]
```

```python
# Now time to make plots
```

```python
plt.hist(Delta_mag_lim, 200)
plt.xlim(-5,0)
plt.gca().invert_xaxis()
plt.title('WDS Histogram of Differential Magnitude')
plt.xlabel('Magnitude of Primary - Magnitude of Secondary')
plt.ylabel('Number of Systems')
plt.savefig('Diffmag_hist_2.png', format ='png', dpi=500)

## There are more recorded binaries of similar magnitude than of different magnitude 
```

```python
x = np.rav(Delta_mag_lim)
y = np.ravel(Sep_last_lim2)

plt.hist2d(x, y, bins = 200, norm=LogNorm())
plt.xlim(xmin = -5, xmax=0)
plt.ylim(ymin = 0, ymax=5.0)
plt.gca().invert_xaxis()
plt.title(' WDS Histogram of Differential Magnitude vs. Separation')
plt.xlabel('Magnitude of Primary - Magnitude of Secondary')
plt.ylabel('Separation (")')
plt.savefig('diffMag_sep_histogram_3.png', format='png',dpi=1000)

## Most oberseved binaries are close together and have similar magnitude
### Why are they close together? 
### (But not completely on top of each other -- .5'')
```
